### 1. Scrape the details of most viewed videos on YouTube from Wikipedia.

### Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos

### You need to find following details:

A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
#connecting to web driver
driver=webdriver.Chrome('chromedriver.exe')

url='  https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
```

```python
#Creating empty list
Rank = []
Name = []
Artist = []
Upload_date = []
Views = []


#Scraping data of rank details
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[1]")[:30]:
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append('-')
    
#Scraping data of video names
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[2]")[:30]:
        Name.append(i.text)
except NoSuchElementException:
    Name.append('-')
    
#Scraping data of artist name
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[3]")[:30]:
        Artist.append(i.text)
except NoSuchElementException:        
    Artist.append('-')
    
#Scraping data of upload dates
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[5]")[:30]:
        Upload_date.append(i.text)
except NoSuchElementException:        
    Upload_date.append('-')
    
#Scraping data of views
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[4]")[:30]:
        Views.append(i.text)
except NoSuchElementException:
    Views.append('-')
    
```

```python
print(len(Rank))              #here I have checked the length of each data so that while creating a dataframe I don't face any error.
print(len(Name)) 
print(len(Artist))
print(len(Upload_date))
print(len(Views))
```

```python
# Creating the DataFrame - 
Most_viewed_videos= pd.DataFrame({'Rank':Rank,' Name':Name,'Artist':Artist,'Upload_date':Upload_date,'Views':Views})
Most_viewed_videos
```

### Q5. Scrape the details of trending repositories on Github.com.
### Url = https://github.com/
### You have to find the following details:

A) Repository title

B) Repository description

C) Contributors count

D) Language used

```python
#connecting to web driver
driver=webdriver.Chrome('chromedriver.exe')

url=' https://github.com/'
driver.get(url)
```

```python
#Options_btn for getting into explore drop down

Options_btn = driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary').click()


# selecting trending 
trending = driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a').click()
time.sleep(5)
```

```python
# creating empty list
repo_urls=[]
title = []
desp = []
count = []
lang = []
language=[]

#Fetching urls for each repository
repository = driver.find_elements_by_xpath("//h1[@class = 'h3 lh-condensed']//a")
for i in repository:
    repo_urls.append(i.get_attribute("href"))


#Scraping data of repository title
ti = driver.find_elements_by_xpath("//h1[@class = 'h3 lh-condensed']")
for i in ti:
    title.append(i.text)

    
#Scraping data from every repository page
for i in repo_urls:
    driver.get(i)
    time.sleep(5)
    
    
    #Scraping data of decription
    try:
        desc = driver.find_element_by_xpath('//p[@class="f4 my-3"]')
        desp.append(desc.text)
    except NoSuchElementException:
        desp.append('-')
        
        
    #Scraping data of contributors
    try:
        cont_tags = driver.find_element_by_xpath("//*[contains(text(),'    Contributors ')]")
        count.append(cont_tags.text.replace('Contributors',''))
    except NoSuchElementException:
        count.append('-')
        
        
    #fetching Languages used
    try:
        for i in driver.find_elements_by_xpath('//a[@class="d-inline-flex flex-items-center flex-nowrap Link--secondary no-underline text-small mr-3"]'):
            lang.append(i.text.replace('\n',''))
        language.append(lang)
    except NoSuchElementException:
        language.append('-')


    
```

```python
print(len(title))              #here I have checked the length of each data so that while creating a dataframe I don't face any error.
print(len(desp)) 
print(len(count))
print(len(language))
```

```python
# Creating the DataFrame - 
Trending_Repo= pd.DataFrame({'Title':title,'Description':desp,'Contributor Counts':count,'Languages Used':language})
Trending_Repo
```

```python
# creating empty list
name=[]
year_span = []
genre = []
runtime = []
rating = []
votes=[]


#Scraping data of  Name details
try:
    for i in driver.find_elements_by_xpath('//h3[@class="lister-item-header"]/a '):
        name.append(i.text)
except NoSuchElementException:
    name.append('-')
    
#Scraping data of Year Span
try:
    for i in driver.find_elements_by_xpath('//span[@class="lister-item-year text-muted unbold"] '):
        year_span.append(i.text)
except NoSuchElementException:
    year_span.append('-')
    
#Scraping data of Genre
try:
    for i in driver.find_elements_by_xpath('//span[@class="genre"]'):
        genre.append(i.text)
except NoSuchElementException:        
    genre.append('-')
    
#Scraping data of runtime
try:
    for i in driver.find_elements_by_xpath('//span[@class="runtime"] '):
        runtime.append(i.text)
except NoSuchElementException:        
    runtime.append('-')
    
#Scraping data of votes
try:
    for i in driver.find_elements_by_xpath('//p[@class="text-muted text-small"][3] '):
        votes.append(i.text.split(":")[1])
except NoSuchElementException:
    votes.append('-')

    
#Scraping data of rating
try:
    for i in driver.find_elements_by_xpath('//div[@class="ipl-rating-star small"]/span[2]'):
        rating.append(i.text)
except NoSuchElementException:
    rating.append('-')
```

```python
# creating empty list
name=[]
typee = []


#Scraping data of  Name details
try:
    for i in driver.find_elements_by_xpath('//p[@class="normal"]/b'):
        name.append(i.text)
except NoSuchElementException:
    name.append('-')
    
#Scraping data of Data Type, Data type, Task , Attribute type, No of instances, No of attribute ,Year and keeping it in single
#list named "typee". we'll extract desirable info in our further steps.
try:
    for i in driver.find_elements_by_xpath('//p[@class="normal"] '):
        typee.append(i.text)
except NoSuchElementException:
    typee.append('-')
    
```

```python
# Keping diffrent required information in diffrent columns with the help of indexing


types=typee[9:4363:7]                  #Datatype

task=typee[10:4363:7]                 # Default Task

atttype=typee[11:4363:7]              # Attribute types

ninstance=typee[12:4363:7]            # Number of instances

natt=typee[13:4363:7]                 # Number of attributes

year=typee[14:4363:7]                 # year
```
