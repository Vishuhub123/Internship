## Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
1. Brand
2. Product Description
3. Price
4. Discount

```python
# connecting to web driver
driver=webdriver.Chrome('chromedriver.exe')
url='https://www.flipkart.com/'
driver.get(url)
```

```python
# selecting search box and writing product we require
search = driver.find_element_by_class_name("_3704LK")
search.send_keys("Sun Glasses")
```

```python
# clicking on search button
btn = driver.find_element_by_class_name('L0Z3Pu')
btn.click()
```

```python
#for extracting data from multiple pages weare using for loop along with exception handling
start=0
end=3

brand = []
price = []
des = []
off = []

for page in range(start,end):
    brands=driver.find_elements_by_class_name('s1Q9rs')
    prices=driver.find_elements_by_xpath("//div[@class='_30jeq3']")
    descp=driver.find_elements_by_class_name('_3Djpdu')
    dis=driver.find_elements_by_class_name('_3Ay6Sb')
    
    for i in brands:
        brand.append(i.text)
    for j in prices:
        price.append(j.text)
    for k in descp:
        des.append(k.text)
    for l in dis:
        off.append(l.text)
    nxt_button=driver.find_elements_by_xpath("//a[@class='_1LKTO3']")
    try:
        driver.get(nxt_button[1].get_attribute('href'))
    except:
        driver.get(nxt_button[0].get_attribute('href'))
        
```

```python
#selecting only 100 products
b=brand[0:100]
p=price[0:100]
d=des[0:100]
o=off[0:100]
```

```python
# creating dataframe
df=pd.DataFrame() 
df['Brand']=b
df['Product Description'] = d
df['Price']=p
df['Discount']=o
df
```

```python
# average salary
average = []
avg=driver.find_elements_by_xpath('//p[@class="averageCtc"]')
for i in avg:
    average.append(i.text)
average
```
